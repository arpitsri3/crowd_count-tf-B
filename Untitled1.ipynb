```python
def heatmap(den, img):
    den_resized = np.zeros((den.shape[0] * 4, den.shape[1] * 4))
    for i in range(den_resized.shape[0]):
        for j in range(den_resized.shape[1]):
            den_resized[i][j] = den[int(i / 4)][int(j / 4)] / 16
    den = den_resized
    #print(den.shape)
    count = np.sum(den)
    print(count)
    data2=np.asarray(den*25)
    #print(data2.shape)
    histogram = np.sum(data2, axis=0)
    histogram[histogram<0] = 0
    histogram = np.around(histogram)
    #print(histogram.shape)
    #print(np.min(histogram))
    #print(np.max(histogram))
    x_range = [np.min(histogram), np.max(histogram)]
    plt.xlabel("Position along the frame width")
    plt.ylabel("Number of People")
    plt.plot(list(range(data2.shape[1])), histogram)
    plt.savefig('./test/count_plot.jpg')
    den = den * 10 / np.max(den)
    w = img.shape[1]
    h = img.shape[0]
    #print(den, '/n')
    #flat=den.flatten()
    #print(flat,'/n')
    #flat.sort()
    #sl = np.int32(len(den)-(len(den)/10))
    #print((w*h), ' ',(len(den)))
    #flat = flat[sl:]
    #print(flat)
    data = []
    print(len(den))
    print(len(den[0]))
    for j in range(len(den)):
        for i in range(len(den[0])):
            for k in range(int(den[j][i])):
                data.append([i + 1, j + 1])

    hm = HeatMap(data, base = './test/img.jpg')
    
    #print(data,'/n')
    return hm
```

```python
def image_resize(image, width = None, height = None, inter = cv2.INTER_AREA):
    # initialize the dimensions of the image to be resized and
    # grab the image size
    dim = None
    (h, w) = image.shape[:2]

    # if both the width and height are None, then return the
    # original image
    if width is None and height is None:
        return image

    # check to see if the width is None
    if width is None:
        # calculate the ratio of the height and construct the
        # dimensions
        r = height / float(h)
        dim = (int(w * r), height)

    # otherwise, the height is None
    else:
        # calculate the ratio of the width and construct the
        # dimensions
        r = width / float(w)
        dim = (width, int(h * r))

    # resize the image
    resized = cv2.resize(image, dim, interpolation = inter)

    # return the resized image
    return resized
```
